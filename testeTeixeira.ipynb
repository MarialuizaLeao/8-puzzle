```python
import copy
from queue import PriorityQueue
import sys
import timeit
import numpy as np
```

```python
class Node:
  def __init__(self, board, parent, cost, heuristic, solution = [board]):
    self.board = copy.deepcopy(board)
    self.parent = parent
    self.cost = cost
    self.heuristic = heuristic
    self.solution = solution
  
  def __lt__(self, other):
    return self.cost + self.heuristic < other.cost + other.heuristic
  
  def __eq__(self, other):
    return self.board == other.board
  
  def __hash__(self):
    return hash(str(self.board))
  
  def __str__(self):
    return str(self.board)

  def getChildren(self):
    children = []
    for action in [moveUp, moveRight, moveLeft, moveDown]:
      newBoard = action(copy.deepcopy(self.board))
      child = Node(newBoard, self, self.cost + 1, 0, self.solution + [newBoard])
      if child.board:
        children.append(child)

    return children
```
